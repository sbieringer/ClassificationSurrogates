# Compare Adam-MCMC and VIB for random-ish in and out of distribution points

```python
n_m_list = len([l for l in os.listdir(m_list_dir) if 'model' in l])
m_list = [f'AdamMCMC_model_{n}.pth' for n in range(n_m_list) if n%100==0]
n_stat = len(m_list)+1
n_stat_alea = 50000

fig,ax = plt.subplots(1,1, figsize = (5,3), sharex = True)#gridspec_kw={'height_ratios': [2,1], 'hspace': 0.})
ax = [ax]

for n_e, event in enumerate([200001, 200003]):#, -100005, -10000, 'pt', 'E', 'n']:
    if isinstance(event, str):
        k = -10000
    else:
        k = event
    i = sort[k]
    conditions = torch.Tensor(dataset_val.features).to(device)
    c = conditions[i:i+1]

    if event == 'pt':
        c[:,1] = (c[:,1]*dataset_val.std_norm[0,1]+500)/dataset_val.std_norm[0,1]
        
    elif event == 'E':
        c[:,4] = (c[:,4]*dataset_val.std_norm[0,4]+3500)/dataset_val.std_norm[0,4]

    elif event == 'n':
        c[:,5] = (c[:,5]*dataset_val.std_norm[0,5]+100)/dataset_val.std_norm[0,5]


    z = torch.randn(n_stat_alea, 1).to(device)

    for j,model in enumerate([model_AdamMCMC, model_VIB]):
        colors_model = colors[j]

        hists = []
        x_gens = []
        for n in tqdm(range(n_stat)):
            if j == 0:
                if n==0:
                    model.load_state_dict(torch.load(save_dir_CFM + f"model_4000.pth"))
                else:
                    model.load_state_dict(torch.load(m_list_dir + m_list[n-1]))
                    #print('loaded model from ' + m_list_dir + m_list[n-1])
            else:
                for layer in model.modules():
                    if isinstance(layer, LinearFlipout):
                        layer.auto_sample = False 
                        layer.sample_weights()

            x_gen = model.decode(z, cond=c.repeat_interleave(n_stat_alea, dim=0))
            
            x_gen = torch.sigmoid(x_gen*20).detach().cpu().numpy()
            #x_gen = (x_gen*20).detach().cpu().numpy()
            x_gens.append(x_gen)
            hist, _ = np.histogram(x_gen, bins = bins)
            hists.append(hist)
        
        x_gens = np.array(x_gens)
        hists = np.array(hists)
        hist_max = np.concatenate([[0], hists.max(0)])
        hist_min = np.concatenate([[0], hists.min(0)])

        #if j == 0:
        #    hist_orig,b,_ = ax[0].hist(x_gens[0],  bins = bins, edgecolor=colors[i], facecolor = None, fill=False, hatch = '/////')
        #else:
        if j == 0:
            hist_orig,b,_ = ax[0].hist(x_gens[0], bins = bins, color=colors_model[n_e], alpha=1., histtype = 'step')
        else:
            hist_orig,b,_ = ax[0].hist(x_gens.mean(0), bins = bins, color=colors_model[n_e], alpha=1., histtype = 'step')

        #bincentres = [-5,0]+[(b[i]+b[i+1])/2. for i in range(len(b)-1)]+[1,5]
        bincentres = b
        ax[0].step(bincentres, hist_max, where='pre',color=colors_model[n_e], alpha = 0.2)
        ax[0].step(bincentres, hist_min, where='pre',color=colors_model[n_e], alpha = 0.2)
        ax[0].fill_between(bincentres, hist_min, hist_max, step = 'pre', color=colors_model[n_e], alpha = 0.5, hatch = hatches[j]
                           , facecolor = 'none' if j == 1 else colors_model[n_e], label=labels[j])

    lower_limit = ax[0].get_ylim()[0]
    upper_limit = 3*ax[0].get_ylim()[1]  
    if not isinstance(event, str):
        x_value = torch.sigmoid(torch.Tensor(dataset_val.target)[i,0]*20).numpy()
        ax[0].vlines(x_value, lower_limit, upper_limit, color = colors_target[n_e], linestyles=':', linewidth = 4)
        
    if n_e == 0:    
        ax[0].text(x_value+0.015, 16, 'ParT \n(QCD jet)', {'color': 'black', 'ha': 'left', 'va': 'top'})
    if n_e == 1:    
        ax[0].text(x_value-0.015, 200, 'ParT \n(top jet)', {'color': 'black', 'ha': 'right', 'va': 'top'})

    #alpha = 0.95
    #quants = np.quantile(x_gens, 1-alpha, axis=1)
    #ax[0].vlines(np.quantile(x_gens.mean(0), 1-alpha), *ax[0].get_ylim(), color = 'red')
    #ax[0].vlines([quants.max(0), quants.min(0)], *ax[0].get_ylim(), color = 'red', linestyle='--')

    ax[0].set_ylabel('occurence')
   #ax[0].set_xlabel(r'$\mathrm{sig}^{-1}$ (ParT/B-CFM output)')
    ax[0].set_xlabel('CFM output')
    ax[0].set_yscale('log')
    ax[0].set_xlim(-0.05,1.05)
    ax[0].set_ylim(None, upper_limit)

legend_handles = [Patch(facecolor='grey', edgecolor='grey', alpha = 0.5,
                         label='VIB-CFM'),
                  Patch(facecolor='none', edgecolor='grey', hatch = 'XXXX',
                         label='AdamMCMC-CFM')]
ax[0].legend(handles = legend_handles, frameon=False)

out_pt = c[0,1]*dataset_val.std_norm[0,1]+dataset_val.means_norm[0,1]
#fig.suptitle(r'$p_T =$' + f'{out_pt:4.1f} GeV')
plt.title('Learned Detector Smearing Distributions')
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].set_ylim(3,99000)

fig.tight_layout()
plt.savefig(f'./figs/data/Learned detector_smearing_distr.pdf')
plt.show()
```

```python
from scipy.spatial.distance import jensenshannon

save_dirs = ['./data/CFM_VIB_sampling_corrected_k100_jets6/', 
             './data/CFM_VIB_sampling_corrected_k50_jets7/', 
             './data/CFM_VIB_sampling_corrected_k50_jets11/', 
             './data/CFM_jets6/',
             './data/CFM_jets7_lrdec0.9999//',
             './data/CFM_jets11_lrdec0.9999//',]
titles = [r'$p_T$', r'$\eta$', r'$\phi$', r'$E_\mathrm{jet}$', r'$n_\mathrm{const}$', ]
bins = np.linspace(0.5,1.1,20)
labels = [r'$p_T$, $\eta$, $\phi$, $E_\mathrm{jet}$, $n_\mathrm{const}$', 
          '+ $m_\mathrm{SD}$',
          r'+ $\tau_1,...,\tau_4$',
          '$p_T$, $\eta$, $\phi$, $E_\mathrm{jet}$, $n_\mathrm{const}$', 
          '+ $m_\mathrm{SD}$',
          r'+ $\tau_1,...,\tau_4$',
          ]
colors_model = ['C0', 'C1', 'C2', 'C0', 'C1', 'C2']

fig, ax = plt.subplots(1, 2, figsize=(5,2.8), sharex='row', sharey='row')

t = torch.sigmoid(torch.Tensor(dataset_val.target)*20)*dataset_val.features[:,0].reshape(-1,1) + (dataset_val.features[:,0].reshape(-1,1)==-1.).astype(float)
hist_parT,_ = np.histogram(t.numpy(), bins = bins)

for j,save_dir in enumerate(save_dirs):

    n_plot = 0 if j<3 else 1

    if j != 1:
        with open(save_dir+'acc.npy', 'rb') as f:
            acc = np.load(f)
    else:
        with open(save_dir+'acc_ep2000.npy', 'rb') as f:
            acc = np.load(f)

    hists = []
    for n_hist in range(acc.shape[1]):
        hist, _ = np.histogram(acc[:,n_hist], bins = bins)
        hists.append(hist)
        
    hists = np.array(hists)
    hist_max = np.concatenate([[0], hists.max(0)])
    hist_min = np.concatenate([[0], hists.min(0)])
    
    jsd = np.zeros(len(hists))
    for i_h, h in enumerate(hists):    
        jsd[i_h] = jensenshannon(hist_parT, h)
    jsd_string = f' JSD = {jsd.mean():4.3f} ± {np.max(np.abs(jsd - jsd.mean())):4.3f}'

    if n_plot == 0:
        hist_orig,b,_ = ax[n_plot].hist(acc[:,0], bins = bins, color=colors_model[j], alpha=1., histtype = 'step')
    else:
        hist_orig,b,_ = ax[n_plot].hist(acc.mean(1), bins = bins, color=colors_model[j], alpha=1., histtype = 'step')

    #bincentres = [-5,0]+[(b[i]+b[i+1])/2. for i in range(len(b)-1)]+[1,5]
    bincentres = b
    ax[n_plot].step(bincentres, hist_max, where='pre',color=colors_model[j], alpha = 0.2)
    ax[n_plot].step(bincentres, hist_min, where='pre',color=colors_model[j], alpha = 0.2)
    ax[n_plot].fill_between(bincentres, hist_min, hist_max, step = 'pre', color=colors_model[j], alpha = 0.5, hatch = hatches[n_plot], 
                            facecolor = 'none' if n_plot == 1 else colors_model[j], 
                            label=labels[j])
    
    print(jsd_string)

    if j==2 or j==5:
        _ = ax[n_plot].hist(t.numpy(), bins = bins, color='red', alpha=1., histtype = 'step', label = 'ParT')
    

ax[0].set_ylabel('occurence')
ax[0].set_title('VIB-CFM')
ax[1].set_title('AdamMCMC-CFM')

for a in ax:
    a.set_yscale('log')
    a.set_xlabel('predicted accuracy')
    a.set_xlim(0.6,1.01)  
    a.set_ylim(5,None)
    
ax[0].legend(handlelength=1, frameon=False)

fig.tight_layout()
fig.subplots_adjust(wspace=.1)

plt.savefig(f'./figs/data/acc_over_input.pdf')
```

```python
#######################
### Define the data ###
#######################

dont_use = ['jet_sdmass', 
            'jet_tau1',	'jet_tau2',	'jet_tau3',	'jet_tau4',
            'aux_genpart_eta', 'aux_genpart_phi', 'aux_genpart_pid', 'aux_genpart_pt', 'aux_truth_match']

dataset = JetDataset("./jet_data",'train', del_context=dont_use)
dataset_val = JetDataset("./jet_data",'val', del_context=dont_use)

np.random.seed(231957)
perm = np.random.permutation(len(dataset_val.features))
n_data = 1000

dataset_val.features = dataset_val.features[perm[:n_data]]
dataset_val.target = dataset_val.target[perm[:n_data]]

save_dirs = ['./data/CFM_VIB_sampling_corrected_k100_jets6/',
             './data/CFM_jets6/']
titles = [r'added $p_T$ [GeV]', r'added $E_\mathrm{jet}$ [GeV]', r'added $n_\mathrm{const}$', ]
n_vals = 10
add_vals = [np.linspace(0,1000,n_vals), np.linspace(0, 4000,n_vals), np.linspace(0,200,n_vals)]
colors_target = ['C4', 'mediumseagreen']
hatches = ['', 'XXXX']  
labels = ['VIB-CFM', 'AdamMCMC-CFM']

fig, ax = plt.subplots(2, 3, figsize=(12,4), sharex='col', sharey='row')

for j,save_dir in enumerate(save_dirs):

    with open(save_dir+'acc_ood.npy', 'rb') as f:
        acc_ood = np.load(f)

    with open(save_dir+'err_diff_ood.npy', 'rb') as f:
        err_diff_ood = np.load(f)

    for i,t in enumerate(titles):
        for n_jet_id, jet_id in enumerate([-1,1]):
            mask = dataset_val.features[:,0]==jet_id
            y = np.mean(acc_ood[i].swapaxes(0,1)[mask],2)
            ax[0,i].plot(add_vals[i], np.median(y,0), color = colors_target[n_jet_id], linestyle = '--' if j == 0 else '-')
            ax[0,i].plot(add_vals[i], np.quantile(y, 0.1, 0), color = colors_target[n_jet_id], alpha = 0.1, linestyle = ':')
            ax[0,i].plot(add_vals[i], np.quantile(y, 0.9, 0), color = colors_target[n_jet_id], alpha = 0.1, linestyle = ':')
            ax[0,i].fill_between(add_vals[i], np.quantile(y, 0.1, 0), np.quantile(y, 0.9, 0), color = colors_target[n_jet_id], alpha = 0.3, label = labels[j],
                                 hatch = hatches[j], facecolor = 'none' if j == 1 else colors_target[n_jet_id])
            
            y = err_diff_ood[i].swapaxes(0,1)[mask]
            ax[1,i].plot(add_vals[i], np.median(y,0), color = colors_target[n_jet_id], linestyle = '--' if j == 0 else '-')
            ax[1,i].plot(add_vals[i], np.quantile(y, 0.1, 0), color = colors_target[n_jet_id], alpha = 0.1, linestyle = ':')
            ax[1,i].plot(add_vals[i], np.quantile(y, 0.9, 0), color = colors_target[n_jet_id], alpha = 0.1, linestyle = ':')
            ax[1,i].fill_between(add_vals[i], np.quantile(y, 0.1, 0), np.quantile(y, 0.9, 0), color = colors_target[n_jet_id], alpha = 0.3, label = labels[j],
                                 hatch = hatches[j], facecolor = 'none' if j == 1 else colors_target[n_jet_id])
            
        ax[1,i].set_xlabel(t)

handles = [Line2D([0], [0], marker='none', linestyle='-', color=colors_target[0], label='QCD jets'),
           Line2D([0], [0], marker='none', linestyle='-', color=colors_target[1], label='top jets'),
           Patch(facecolor='grey', edgecolor='grey', alpha = 0.5,
                         label='VIB-CFM'),
           Patch(facecolor='none', edgecolor='grey', hatch = 'XXXX',
                    label='AdamMCMC-CFM')
                    ]

ax[1,0].legend(handles = handles, frameon= False, loc='center right')

for i in range(2):
    ax[i,0].set_xlim(0,1000)
    ax[i,1].set_xlim(0,2500)
    ax[i,2].set_xlim(0,200)

for a in ax:
    for ai in a:
        ai.grid(alpha = 0.5)
    
lim_add = 0.1
ax[0,0].set_ylim(-lim_add,1 + lim_add)
ax[0,1].set_ylim(-lim_add,1 + lim_add)
ax[0,2].set_ylim(-lim_add,1 + lim_add)
ax[1,0].set_ylim(-lim_add,1 + lim_add)
ax[1,1].set_ylim(-lim_add,1 + lim_add)
ax[1,2].set_ylim(-lim_add,1 + lim_add)


ax[0,0].set_ylabel('predicted accuracy')
ax[1,0].set_ylabel('epistemic uncertainty')

fig.tight_layout()
fig.subplots_adjust(wspace=.08)

plt.savefig(f'./figs/data/ood.pdf')
```
